# Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
import numpy as np

ratings = pd.read_csv('ml-latest-small/ratings.csv')

ratings_count = ratings.groupby('userId').count()
ratings_count = ratings_count[['rating']]
ratings_count.rename(columns={'rating':'count_rating'}, inplace=True)
ratings100 = ratings_count.query('count_rating >= 100')


ratings_grouped = ratings.groupby('userId').agg({'timestamp' : [min, max]})
ratings_grouped['life'] = ratings_grouped['timestamp']['max'] - ratings_grouped['timestamp']['min']
ratings_life = ratings_grouped[['life']]

ratings_joined = ratings100.join(ratings_life)
ratings_joined.mean()
```
